## LOOPS

```julia
n=1
while n<10
    n+=1
    println(n)
end    
```

```julia
for n in 1:10
      println(n)
end    
```

```julia
for n= 1:10
     println(n)
end    
```

```julia
m,n=5,5
A= zeros(m,n)
A
```

```julia
for i in 1:m
    for j in 1:n
        A[i,j]=i+j
        
    end
end 
A
```

```julia
#Instead of the above we can use the below syntax 
B=zeros(m,n)
for i in 1:m, j in 1:n
    B[i,j]=i+j
end
B
```

```julia
c=[i+j for i in 1:m, j in 1:n]
display(c)
```

## Conditional statements

```julia
x=10
y=4
if x>y
    println("x>y")
elseif y>x
    println("y>x")
else
    println("x=y") 
end    
```

### Mutating vs non-mutating functions

```julia
#By convention, functions folllowed by ! alter their contents and functions lacking ! do not

v=[3,5,2]
sort(v)
print(v)
```

```julia
#Above v and sort(v) are not same
sort!(v)
println(v)
```

### Broadcasting

```julia
A = [i + 3*j for j in 0:2, i in 1:3]
display(A)
f(A)
```

```julia
f.(A)
```

```julia
v=[1,2,3]
f(v)
```

```julia
f.(v)
#Like it is operation over individual element
```
